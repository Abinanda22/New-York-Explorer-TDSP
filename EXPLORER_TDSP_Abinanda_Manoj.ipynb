**Step 2:** Let's begin by creating a chart that displays the average number of crashes per hour of the day. This will help us understand whether additional factors are contributing to crashes - i.e. rush hour, school dismissal time, night fall, etc.

````python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load the dataset
file_path = "/content/drive/MyDrive/ExplorerTDSP:AbinandaManoj/Motor_Vehicle_Collisions_-_Crashes_20250118.csv"
data = pd.read_csv(file_path)

# Convert 'CRASH DATE' and 'CRASH TIME' to datetime
data['CRASH DATE'] = pd.to_datetime(data['CRASH DATE'])
data['CRASH TIME'] = pd.to_datetime(data['CRASH TIME'], format='%H:%M')

# Time of Day Analysis
data['Hour of Day'] = data['CRASH TIME'].dt.hour

# Group by 'Hour of Day' and calculate the average number of crashes per hour
average_crashes_per_hour = data.groupby('Hour of Day').size() / data['Hour of Day'].nunique()

# Plot the average number of crashes
plt.figure(figsize=(12, 6))
sns.barplot(x= average_crashes_per_hour.index , y= average_crashes_per_hour.values)
plt.title('Average Number of crashes per Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Average Number of crashes')
plt.xticks(range(0, 24))
plt.show()
````

**Step 4**: Apply time series decomposition to review trends, seasonality, and residuals. New to time series analysis? Review the [Time Series Flashcard video series](https://youtube.com/playlist?list=PLNs9ZO9jGtUAqd0CNKySksPJJaOqYPMvk&feature=shared) here to learn about trends, components, and further analysis!

````python
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose

# Count the number of crashes per day, group by CRASH DATE
daily_crashes = data.groupby('CRASH DATE').size()

# Set plot style
sns.set(style="darkgrid")

# Plot the daily crashes time series
plt.figure(figsize=(15, 6))
plt.plot(daily_crashes, label='Daily crashes')
plt.title('Daily Motor Vehicle Collisions in NYC')
plt.xlabel('Day')
plt.ylabel('Number of crashes')
plt.legend()
plt.show()

# Decompose the time series
decomposition = seasonal_decompose(daily_crashes, model='additive', period=365)

# Plot the decomposed components
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(15, 12))
decomposition.trend.plot(ax=ax1)
ax1.set_title('Trend')
decomposition.seasonal.plot(ax=ax2)
ax2.set_title('Seasonality')
decomposition.resid.plot(ax=ax3)
ax3.set_title('Residuals')
plt.tight_layout()
plt.show()
````

The visualizations above provide valuable insights into the time series of daily motor vehicle collisions in New York City:

1. Time Series Plot: This shows the number of daily crashes over time. You might observe long-term trends, seasonal patterns, or significant outliers.

2. Decomposed Components:
  
    2.1 Trend: This graph shows the long-term trend in the data, which can indicate whether crashes are increasing, decreasing, or stable over time.

    2.2 Seasonality: This reveals any regular patterns that repeat over a specific period, such as yearly. It helps identify times of the year with higher or lower crash frequencies.

    2.3 Residuals: These are the irregular components that cannot be attributed to the trend or seasonality. They might include random or unpredictable fluctuations.

````python
#TODO: Continue building your heatmap
# Sample a subset of the data for visualization
sample_data_severity = data_geo.sample(n=1000, random_state=42)

# Create a base map
m_severity = folium.Map(location=[40.730610, -73.935242], zoom_start=10)

# Add crashes to the map with color coding and shape coding based on severity
for index, row in sample_data_severity.iterrows():
    if row['NUMBER OF PERSONS KILLED'] > 0:
        color = "RED"  # Fatalities

        folium.features.RegularPolygonMarker(
          location=[row['LATITUDE'], row['LONGITUDE']],
          number_of_sides=3,
          radius=5,
          gradient = False,
          color=color,
          fill=True,
          fill_color=color
        ).add_to(m_severity)


    elif row['NUMBER OF PERSONS INJURED'] > 0:
        color = "BLUE"  # Injuries
        folium.CircleMarker(
          location=[row['LATITUDE'], row['LONGITUDE']],
          radius=5,
          color=color,
          fill=True,
          fill_color=color
       ).add_to(m_severity)
    else:
        color = "GRAY"  # No injuries or fatalities
        folium.features.RegularPolygonMarker(
          location=[row['LATITUDE'], row['LONGITUDE']],
          number_of_sides=4,
          radius=5,
          gradient = False,
          color=color,
          fill=True,
          fill_color=color
        ).add_to(m_severity)


m_severity.save("severity.html")
````

````python
# Change 'Boro' column entries to uppercase
TrafficData['Boro'] = TrafficData['Boro'].str.upper()
#Standardize column names
TrafficData = TrafficData.rename(columns={'Boro': 'BOROUGH'})
# Merge 'M', 'D', and 'Yr' columns into 'TrafficDate' in the format m/d/yr
TrafficData['TrafficDate'] = TrafficData.apply(lambda row: f"{row['M']}/{row['D']}/{row['Yr']}", axis=1)
# Merge 'HH' and 'MM' columns into 'TrafficTime' in the format hh:mm
TrafficData['TrafficTime'] = TrafficData.apply(lambda row: f"{row['HH']}:{row['MM']}", axis=1)
# Drop the original 'M', 'D', and 'Yr' columns
TrafficData = TrafficData.drop(['M', 'D', 'Yr','HH', 'MM'], axis=1)
````

````python
trafficData = TrafficData[['BOROUGH', 'TrafficDate', 'TrafficTime', 'Vol']]
# Check the updated data
trafficData.head()
````

````python
CrashData = data[['BOROUGH', 'CRASH DATE', 'CRASH TIME', 'LATITUDE', 'LONGITUDE']]
crashData = CrashData.dropna()
crashData.head()
````

````python
#Standardize column names
trafficData = trafficData.rename(columns={'TrafficDate': 'CRASH DATE', 'TrafficTime': 'CRASH TIME'})


#Data Preprocessing: Merge datasets by borough and datetime
merged_data = pd.merge(
    crashData,
    trafficData,
    on=['BOROUGH', 'CRASH DATE', 'CRASH TIME'],
    how='inner'
)


merged_data.info()
````

````python
# Aggregate traffic volume by borough
borough_traffic_volume = merged_data.groupby('BOROUGH')['Vol'].sum().reset_index()

# Create a barplot for total traffic volume by borough
plt.figure(figsize=(12, 6))
sns.barplot(data=borough_traffic_volume, x='BOROUGH', y='Vol', palette='viridis')

# Add title and labels
plt.title('Total Traffic Volume by Borough')
plt.xlabel('Borough')
plt.ylabel('Total Traffic Volume')

# Show the plot
plt.xticks(rotation=45)
plt.show()
````

````python
# Ensure CRASH TIME is in datetime format and extract the hour
merged_data['CRASH TIME'] = pd.to_datetime(merged_data['CRASH TIME'], format='%H:%M')
merged_data['Hour of Day'] = merged_data['CRASH TIME'].dt.hour

# Group by 'BOROUGH' and 'Hour of Day' to calculate the average traffic volume
borough_hourly_traffic = merged_data.groupby(['BOROUGH', 'Hour of Day'])['Vol'].mean().reset_index()

# Group by 'BOROUGH' and 'Hour of Day' to calculate the average number of crashes
borough_hourly_crashes = merged_data.groupby(['BOROUGH', 'Hour of Day']).size().reset_index(name='total_crashes')
borough_hourly_crashes['average_crashes'] = borough_hourly_crashes['total_crashes'] / merged_data['BOROUGH'].nunique()

# Plotting
boroughs = borough_hourly_traffic['BOROUGH'].unique()

for borough in boroughs:
    # Create a figure
    plt.figure(figsize=(12, 6))

    # Filter the data for the specific borough
    traffic_data = borough_hourly_traffic[borough_hourly_traffic['BOROUGH'] == borough]
    crashes_data = borough_hourly_crashes[borough_hourly_crashes['BOROUGH'] == borough]

    # Plot the average traffic volume as a barplot (no legend for this plot)
    sns.barplot(data=traffic_data, x='Hour of Day', y='Vol', palette='viridis', alpha=0.6)

    # Plot the average number of crashes as a line plot (with legend)
    sns.lineplot(data=crashes_data, x='Hour of Day', y='average_crashes', color='red', marker='o', label='Avg Crashes')

    # Set the plot labels and title
    plt.title(f'Average Traffic Volume and Number of Crashes per Hour of Day in {borough}')
    plt.xlabel('Hour of Day')
    plt.ylabel('Values')
    plt.xticks(range(0, 24))

    # Show legend only for the crashes line
    plt.legend(loc='upper right')

    # Display the plot
    plt.show()
````

````python
#Correlation analysis
# Group by 'BOROUGH' and 'Hour of Day' to calculate the total crashes and average traffic volume
borough_hourly_patterns = merged_data.groupby(['BOROUGH', 'Hour of Day']).agg(
    total_crashes=('CRASH DATE', 'count'),
    avg_traffic_volume=('Vol', 'mean')
).reset_index()

# Identify boroughs with the highest total crashes
peak_crash_boroughs = borough_hourly_patterns.groupby('BOROUGH')['total_crashes'].sum().nlargest(5).index
# Filter for boroughs with the highest crash frequencies
peak_borough_data = borough_hourly_patterns[borough_hourly_patterns['BOROUGH'].isin(peak_crash_boroughs)]

# Calculate the correlation between crashes and traffic volume for each borough
correlations = peak_borough_data.groupby('BOROUGH').apply(
    lambda group: group[['total_crashes', 'avg_traffic_volume']].corr().iloc[0, 1]
).reset_index(name='correlation')

# Display the correlation for each borough
print("Correlation between crashes and traffic volume for top crash boroughs:")
print(correlations)
````

````python
# Create a barplot for Correlation between crashes and traffic volume for top crash boroughs
plt.figure(figsize=(12, 6))
plot_data = pd.merge(peak_borough_data, correlations, on='BOROUGH')
sns.barplot(data=plot_data, x='BOROUGH', y='correlation', palette='viridis')

# Add title and labels
plt.title('Correlation between crashes and traffic volume for top crash boroughs')
plt.xlabel('Borough')
plt.ylabel('Correlation Rate')

# Show the plot
plt.xticks(rotation=45)
plt.show()
````

````python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Assuming `merged_data` does not have 'total_crashes', calculate it:
merged_data['total_crashes'] = merged_data.groupby(['BOROUGH', 'Hour of Day'])['CRASH DATE'].transform('count')

# Pivot the table to create a feature matrix
hourly_pattern = merged_data.groupby(['BOROUGH', 'Hour of Day'])['total_crashes'].sum().unstack().fillna(0)

# Normalize the data
scaler = StandardScaler()
hourly_pattern_scaled = scaler.fit_transform(hourly_pattern)

# Convert back to DataFrame for ease of use
hourly_pattern_scaled = pd.DataFrame(hourly_pattern_scaled, index=hourly_pattern.index, columns=hourly_pattern.columns)
````

````python
from sklearn.metrics import silhouette_score

# Assuming 'borough_hourly_patterns' contains features like 'total_crashes' and 'avg_traffic_volume'
features = borough_hourly_patterns[['total_crashes', 'avg_traffic_volume']].values
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

silhouette_scores = []

for k in range(2, 11):  # Silhouette is not defined for 1 cluster
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_features)
    score = silhouette_score(scaled_features, kmeans.labels_)
    silhouette_scores.append(score)

# Plot the Silhouette Scores
plt.figure(figsize=(8, 5))
plt.plot(range(2, 11), silhouette_scores, marker='o')
plt.title("Silhouette Scores for Optimal k")
plt.xlabel("Number of Clusters")
plt.ylabel("Silhouette Score")
plt.show()
````

````python
optimal_k = 3  # Based on the silhouette method
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
kmeans.fit(scaled_features)
borough_hourly_patterns['KMeans_Cluster'] = kmeans.labels_
````

````python
from scipy.cluster.hierarchy import dendrogram, linkage

# Linkage matrix
linkage_matrix = linkage(scaled_features, method='ward')  # Ward's method minimizes variance
# Plot dendrogram
plt.figure(figsize=(10, 7))
dendrogram(linkage_matrix, labels=borough_hourly_patterns['BOROUGH'].values, leaf_rotation=90)
plt.title("Dendrogram for Hierarchical Clustering")
plt.xlabel("Boroughs")
plt.ylabel("Distance")
plt.show()
````

````python
from scipy.cluster.hierarchy import fcluster
import seaborn as sns

# Define the number of clusters based on the dendrogram
hierarchical_clusters = fcluster(linkage_matrix, t=optimal_k, criterion='maxclust')  # 3 clusters
borough_hourly_patterns['Hierarchical_Cluster'] = hierarchical_clusters

# K-Means Results
sns.scatterplot(data=borough_hourly_patterns, x='total_crashes', y='avg_traffic_volume', hue='KMeans_Cluster', palette='viridis')
plt.title("K-Means Clustering")
plt.xlabel("Total Crashes")
plt.ylabel("Average Traffic Volume")
plt.show()

# Hierarchical Clustering Results
sns.scatterplot(data=borough_hourly_patterns, x='total_crashes', y='avg_traffic_volume', hue='Hierarchical_Cluster', palette='viridis')
plt.title("Hierarchical Clustering")
plt.xlabel("Total Crashes")
plt.ylabel("Average Traffic Volume")
plt.show()
````

````python
import statsmodels.api as sm

# Group by 'BOROUGH' and 'Hour of Day' to calculate the total crashes and average traffic volume
borough_hourly_patterns = merged_data.groupby(['BOROUGH', 'Hour of Day']).agg(
    total_crashes=('CRASH DATE', 'count'),
    avg_traffic_volume=('Vol', 'mean')
).reset_index()

# Filter for boroughs with the highest crash frequencies
peak_crash_boroughs = borough_hourly_patterns.groupby('BOROUGH')['total_crashes'].sum().nlargest(5).index
peak_borough_data = borough_hourly_patterns[borough_hourly_patterns['BOROUGH'].isin(peak_crash_boroughs)]

# Run Linear Regression for each borough
for borough in peak_borough_data['BOROUGH'].unique():
    borough_data = peak_borough_data[peak_borough_data['BOROUGH'] == borough]

    # Define the independent variable (traffic volume) and dependent variable (total crashes)
    X = borough_data['avg_traffic_volume']
    y = borough_data['total_crashes']

    # Add a constant to the independent variable (for the intercept)
    X = sm.add_constant(X)

    # Fit the Poisson regression model
    poisson_model = sm.GLM(y, X, family=sm.families.Poisson()).fit()

    # Print the summary of the regression results
    print(f"Poisson Regression results for {borough}:")
    print(poisson_model.summary())
    print("\n")
    # Predicted values
    borough_data['predicted_crashes'] = poisson_model.predict(X)

    # Plot observed vs predicted
    plt.figure(figsize=(10, 6))
    plt.scatter(borough_data['avg_traffic_volume'], borough_data['total_crashes'], label='Observed', alpha=0.7)
    plt.scatter(borough_data['avg_traffic_volume'], borough_data['predicted_crashes'], label='Predicted', alpha=0.7, color='red')
    plt.title(f"Observed vs Predicted Crashes in {borough}")
    plt.xlabel('Average Traffic Volume')
    plt.ylabel('Total Crashes')
    plt.legend()
    plt.show()
````
